```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt
from pathlib import Path
Path.ls = lambda x: list(x.iterdir())
```

```python
p_data = Path()/'dataset/picture/train'
```

```python
plt.imshow(plt.imread(p_data.ls()[6]))
```

```python
img = cv.imread(str(p_data.ls()[6]),3)
```

```python
plt.imshow(img)
```

```python
edges = cv.Canny(img, 40,100)

fig, axes = plt.subplots(ncols=2, figsize=[15,7])
ax1, ax2 = [*axes.flatten()]
ax1.imshow(img)
ax2.imshow(edges, cmap='gray')
plt.tight_layout()
```

```python
(edges/255).mean()
```
